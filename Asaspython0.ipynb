```python
def hitung_biaya_ekspedisi(panjang, lebar, tinggi, berat):
  volume = panjang * lebar * tinggi
  biaya_dasar = 5000 * berat

  if volume > 50 * 50 * 50:
    biaya_tambahan = 2000 * berat
  else:
    biaya_tambahan = 0

  total_biaya = biaya_dasar + biaya_tambahan
  return total_biaya

# Contoh penggunaan
panjang = float(input("Masukkan panjang paket (cm): "))
lebar = float(input("Masukkan lebar paket (cm): "))
tinggi = float(input("Masukkan tinggi paket (cm): "))
berat = float(input("Masukkan berat paket (kg): "))

biaya = hitung_biaya_ekspedisi(panjang, lebar, tinggi, berat)
print("Total biaya pengiriman:", biaya)
```
